```python
from pyquery import PyQuery as pq
import re

a = pq('https://www.18onlygirlsblog.com/category/movies/')

items = a('#content li.box-shadow')

for item in items.items():
    print(item('img').attr('src'))
    print(item('img').attr('srcset'))
    print(item('a').attr('href'))
    print(item('a').attr('title'))
    
    result = re.findall('[a-zA-z]+://[^\s]*', item('img').attr('srcset'))
    print(result)

    
    b = pq(item('a').attr('href'))
    print(b('video source').attr('src'))
    print(b('video a').attr('href'))

    previews = b('.ngg-gallery-thumbnail a')
    for preview in previews.items():
        print(preview('a').attr('data-src'))
        print(preview('a').attr('data-thumbnail'))
    
    print('\n')
    
```

```python
import requests
import re
from requests.exceptions import RequestException
from multiprocessing import Pool
from bs4 import BeautifulSoup
from urllib.parse import urljoin
import json
from pyquery import PyQuery as pq
import os,sys

url = 'https://xartfan.com/x_art-aliyah-in-sexual-vanity/'
b = pq(url)

video = None
player = b('div.flowplayer')
if player:
    src = json.loads(player.attr('data-item')).get('sources')[0].get('src')     
    board = re.search('background-image: url\((.*?)\)', player.attr('style')).group(1)
    video = {
        'src':src,
        'board':board
        }

previews = b('div.tiled-gallery-item a')
details = []
for preview in previews.items():
    details.append({
        'large': preview('img').attr('data-large-file'),
        'mid': preview('img').attr('data-medium-file'),
        'small': preview('img').attr('src'),                    
        }
        )


image = {           
    'url':  url,
    'video': video,
    'detail':details,
    }

print(image)
```
